
Do this in the following ways:

````python

u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

#a)
prod = np.outer(u,v)
print('#a')
print(prod)

#b)
prod = np.array([[x*y for y in v] for x in u]).reshape(4,4)
print('#b')
print(prod)

#c
u = u.reshape(len(u), 1)
print('#c')
prod = u*v
print(prod)
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
N = 10000

def prime_search(N):
    mask = np.array([True for i in range(0,N+1)])

    for num in range(2, N+1):
        for i in range(num+1,N+1):
            if mask[i] == True:
                if i % num == 0:
                    mask[i] = False
    #print(num, mask[num])
    
    numbers = np.arange(0,N+1)
    primes = []

    for i in range(0,N+1):
        if mask[i] == True:
            primes += [i]
    return primes

#Checking the performance
import time
times = []
for N in [10,100,500,1000,5000,10000]:
    tick = time.time()
    prime_search(N)
    tock = time.time()
    deltat = tock - tick
    print('Time for N =', N, ':', deltat)
    times += [deltat]

````

````python
import matplotlib.pyplot as plt
N_array = [10,100,500,1000,5000,10000]
plt.figure()
plt.plot(N_array, times, marker = 'o')
plt.grid()
plt.xlabel('N')
plt.ylabel('Time')
#The time performance scales exponentially
````

````python
def sieve_eratosthenes(N):
    A = [True for i in range(0,N)]
    
    for i in range(2,N+1):
        if i < N**(0.5):
            if A[i] == True:
                j = i**2
                while j < N:
                    A[j] = False
                    j += i
                    
    numbers = np.arange(0,N+1)
    primes = []

    for i in range(0,N):
        if A[i] == True:
            primes += [i]
    return primes
    
print(sieve_eratosthenes(5000))
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
N = 1000
M = 200
grid = np.random.randint(0,2,size = (N,M))

#Array of -1 and 1
grid = grid*2 -1


distance = np.zeros((N,M))
for i in range(0,N):
    for j in range(1,M):
        distance[i,j] = distance[i,j-1] + grid[i,j]
````

````python
distancesq = distance**2

average = np.sqrt([[sum(distancesq[:,j])/N] for j in range(0,M)])
````

````python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(np.arange(0,200), average)
plt.grid()
plt.ylabel('Average distance in function of the step')
plt.xlabel('Step')
````

I was expecting an average distance equal to zero since the left and right step have the same probability.

````python
filename = 'populations.txt'
data = np.genfromtxt(filename, skip_header = 1)
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]
carrots = data[:,3]

plt.figure()
plt.plot(years, hares, label = 'Hares', marker = 'o')
plt.plot(years, lynxes, label = 'Lynxes', marker = 'o')
plt.plot(years,carrots, label = 'Carrots', marker = 'o')
plt.xlabel('Years')
plt.grid()
plt.legend()

#Main properties

print('Mean of hares population = ', hares.mean())
print('Mean of lynxes population = ', lynxes.mean())
print('Mean of carrots population = ', carrots.mean())

print('Standard deviation for hares = ', hares.std())
print('Standard deviation for lynxes = ', lynxes.std())
print('Standard deviation for carrots = ', carrots.std())
````
